```
import pydot
from collections import defaultdict 
from IPython.display import Image, display
```

```
def lstr(p):
        return ''.join([str(elem)  for elem in p])
```

```
def ModifiedBFS(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def ModifiedBFSPreSolsLevel(s,L):  

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
            if flag:
              node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
            else: 
              node = pydot.Node(lstr(s))

            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)) and (l < L):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                     node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def voltear_intervalo(lista,inicio,fin):
  l = len(lista)
  corte_intacto1 = lista[0:inicio]
  corte_intacto2 = lista[(fin+1):l]

  seccion_reversada = lista[inicio:(fin+1)]
  seccion_reversada.reverse()
  #print(corte_intacto1,seccion_reversada,corte_intacto2)
  sol = corte_intacto1 + seccion_reversada+ corte_intacto2
  return sol
```

#**Metodo para graficar una red de pancakes**

Esta funcion surge de modificar la funcion  `ModifiedBFS(s)`. La forma en que funciona es que por cada nodo se obtienen sus vecinos directos y se realiza la coneccion, luego se realiza la misma operacion con cada nodo descubierto. El algoritmo se detiene cuando los nodos hallados como vecinos ya hayan sido todos descubiertos previamente.

```
def Red_de_pancakes(s): 

        aristas = []
      #__________________________________________________________________
      #Inicializacion del algoritmo

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="graph")
        
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True

       #__________________________________________________________________

       #Nodo a buscarle nodos conectados
        while queue: 
          
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)

            visitedN = defaultdict()
            visitedN[lstr(s)] = True
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            #if(s != sorted(s)):
            for i in range(len(s)-1):

              #Flip
              d = s[i+1::-1] + s[i+2:]  
              
              #si es nuevo agregarlo
              if (d != s) and not(lstr(d) in  visitedN.keys()):
                
                #___________________________________________________________________
                #Color del nodo
                if(d != sorted(d)):
                  node = pydot.Node(lstr(d))
                else:
                  node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                #___________________________________________________________________
                
                if(not(lstr(d) in  visited.keys())):
                  #print("Nodos usados ",visited.keys())
                  #print("Nodos conectados n ", visitedN.keys())
                  queue.append(d)     
                  #print(queue) 

                G.add_node(node)
                edge = pydot.Edge(lstr(s),lstr(d))
      
                #Actualizar la lista de aristas dibujadas 
                aristas.append(lstr(s) + lstr(d))
                #print("agregado ",lstr(s) + lstr(d))
                
                #si ya se dibujo la arista, no dibujarla nuevamente
                if(not(lstr(d) + lstr(s)) in aristas):
                  #print("Oh")   
                  G.add_edge(edge)
                  
                graph[lstr(s)].append(lstr(d))
                
                visited[lstr(d)] = True 
                visitedN[lstr(d)] = True 

        im = Image(G.create_png())
        display(im)
```

#**Funcion para obtener el minimo numero de flips necesarios para ordenar una pila de pancakes especifica**

Esta funcion surge de modificar la funcion  `ModifiedBFSPreSolsLevel(s,L)` la cual usa Breadth First Search.

La idea parte de lo siguiente
1. El grafo que se va construyendo incluye todas las formas en que se pueden efectuar flips a la pila de pancakes
2. La profundiadad de un nodo en el grafo equivale a los flips realizados para llegar a el desde el nodo raiz

Cuando el algoritmo esta hallando los hijos de un nodo en el nivel l, que se ubican en el nivel  l+1, y se verifica que alguno estos hijos ya esta ordenado, entonces retornamos la profundidad del nodo ordenado descubierto, es decir l+1

```
def Pn_stack_especifico(s):  

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]

            node = pydot.Node(lstr(s))

            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green 
                    return (l+1)
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 

        return 0 
```

#**Funcion para hallar $Pn$ (minimo numero de flips necesarios para ordenar peor caso)**

La siguiente funcion funciona de la siguiente manera:
1. Obtener una lista con los enteros de 1 a n
2. A partir de la lista anterior generar las n! permutaciones posibles. Se genera una lista de listas que guarda todas las pilas posibles.
3.Se itera por todos los casos calculando el menor numero de flips para ordenar cada uno. Se halla el maximo y se retorna como respuesta. 

```
def Hallar_Pn (n):

  lista = obtener_lista_1_a_n(n)
  #print("lista ordenada ",lista)

  Espacio_a_recorrer = list(permutaciones(lista))
  #print(list(Espacio_a_recorrer))
  
  numero_permutaciones = len(Espacio_a_recorrer)
  minimos_flips_peor_caso = 0

  for i in range(0,numero_permutaciones):
    caso_i = list(Espacio_a_recorrer[i])
    #print("lista caso i ",caso_i)
    flips_caso_i = Pn_stack_especifico(caso_i)
    #print("flips caso i ",flips_caso_i)
    minimos_flips_peor_caso = max(flips_caso_i,minimos_flips_peor_caso)

  return minimos_flips_peor_caso
```

#**Metodo para graficar arboles de pancakes usando 2 espatulas**

Este metodo va construyendo un grafo donde en cada paso se descubren todos los hijos de un nodo, que corresponden a permutaciones a las que se llega en 1 flip con 2 espatulas desde el nodo actual. Cuando a todos los nodos posibles ya se les hayaron sus nodos hijos y se agregaron al grafo, omitiendo repeticiones, el algoritmo termina.

Los nodos hijos de un nodo salen de todos los resultados de hacer el flip de cualquier intervalo de la pila actual.Los intervalos son hallados usando combinaciones y los flips se hacen usando una funcion que recibe la lista y el intervalo a invertir.

```
def Arbol_con_dos_espatulas(s): 

        #__________________________________________________________
        #Inicializacion del algoritmo

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
        #__________________________________________________________
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            #print("revisando hijos de ",s)
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 

            
            if(s != sorted(s)):
              #Obtener posibles intervalos para hacerle flip a la pila 
              pf = posibles_flips((len(s) - 1))
              #print("C = ",pf)
              
              nf = len(pf)
              for i in range (nf):
                intervalo_dos_espatulas = pf[i]
                d = voltear_intervalo(s,intervalo_dos_espatulas[0], intervalo_dos_espatulas[1])
                #print("flip del intervalo ", intervalo_dos_espatulas,"es ",d)

                if (d != s) and not(lstr(d) in visited.keys()):

                  if (d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
                  #print("visitados ",visited)
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```
